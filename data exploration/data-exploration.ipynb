```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
```

```python
data = pd.read_csv('caravan-insurance-challenge.csv')

profile=ProfileReport(data, explorative=False, pool_size=4, progress_bar=True ,minimal=True) # Had to use minimal to be under github max file size for the report
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize = (9,11))
data[data['CARAVAN']==0].groupby('APERSAUT').APERSAUT.count().plot(kind='bar',title='number of car policy distribution for noncustomers', ax=axes[0], label='Caravan=0')
data[data['CARAVAN']==1].groupby('APERSAUT').APERSAUT.count().plot(kind='bar',title='number of car policy distribution for customers', ax=axes[1], label='Customer')
plt.show()
```
